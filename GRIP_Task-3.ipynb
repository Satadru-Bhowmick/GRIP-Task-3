# Analysis of Sales vs Region

```python
fig, ax=plt.subplots(figsize=(8,5))
x=data['Region']
y=data['Sales']
ax.bar(x,y)
plt.show()
```

Maximum sale is in south region whereas east has the minimum

# Analysis of Region vs Profit

```python
fig, ax=plt.subplots(figsize=(8,5))
x=data['Region']
y=data['Profit']
ax.bar(x,y)
plt.show()
```

Maximum Profit is earned from the Central region whereas East region gives the min profit

# Analysis of State vs Sales

```python
fig, ax=plt.subplots(figsize=(17,8))
x=data['State']
y=data['Sales']
ax.bar(x,y)
plt.xticks(rotation=90)
plt.show()
```

Florida has the highest Sales whereas Kansas has the lowest.

# Analysis of State vs Profit

```python
fig, ax=plt.subplots(figsize=(17,8))
x=data['State']
y=data['Profit']
ax.bar(x,y)
plt.xticks(rotation=90)
plt.show()
```

There are few states where profit is very less or minimum

# Analysis of Sales vs Category

```python
fig, ax=plt.subplots(figsize=(8,5))
x=data['Category']
y=data['Sales']
ax.bar(x,y)
plt.show()
```

Technology sector has the higest amount of Sales whereas Furniture has the lowest

# Analysis of Category vs Profit
